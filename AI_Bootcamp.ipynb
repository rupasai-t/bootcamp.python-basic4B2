```python
#step-2: grouping the input columns and output columns and convert the tablier to array formet
x=df.iloc[:,0:3]#slicing the input column
```

```python
# step-3: Handing missing data or Nalivalues.
# scenario-1: missing values of numerical column.
#we have calculate the average of the column replace the average or mean value or empty values.
# scenario-2:missing value in numerical colume.
#fill the value we have replace the empty  values with high repeating values.
# scenario-3: haveing missing values  in huge data set.
#remove those empty values from the data set.

import numpy as np
from sklearn.impute import SimpleImputer #simpleimputer is a class from sklearn .
#which is used to new values placed in the missing values.simpleimputer is a sklearn are packages.
#simpleimputer is a class which  was imported from sklearn package.

imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(x[:,1:3])
x[:,1:3]=imputer.transform(x[:,1:3])
```

```python
x
```

```python

#step-4: handling the categorical data
#converting the categorical data into numerical data.
#OneHotencoder is a class which was imported from sklearn package.
#OneHotencoder is used to convert text to binarycode(0,1).
#columntransformer is a class which was to import from sklearn package.
#columntransformer is used  to apply changes or transformer(updating the values present in the data is transformer)
#columntransformer is a class which is used to apply the changes of numeric data in the place of categorical data
#Alexa-0,0,1

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
x=ct.fit_transform(x)


```

```python
x
```

```python
#Onehotencoder is only form input column
#linearencoder is only for output column

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
#step-5: split the data into training and testing data
#train and test split is a class
#which was imported from sklearn.model_selection.
#it was used for spliting the data into training and testing.

from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#step-6:feature scaleing
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train[:,3:]=sc.fit_transform(x_train[:,3:])
x_test[:,3:]=sc.transform(x_test[:,3:])
```

```python
#machine learning
#supervisioned learning - when we have data  and  the model tast are  classification are prediction.
#unsupervisioned learning - unlabel  data, model remembers the patterns. we can perform prediction also.
#rainforcement learning - model will learn form falilures. trail and error method.
#supervisioned learnig is two types
#logistic regression -when the model perform  classification tast,
#we have to apply logistic reggration algorithm(eg:-yes or no,0's and 1's ,true or false).
#linear regression -  when the model perform prediction task,
#we have to apply linear reggration algorithm(eg:-  label or model  prediction).
#step-7:  apply algorithm to the training data for model training
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)

```

```python
y_pred=classifier.predict(x_test)
```
